```python
def plot_em(emcts, snr, order, mdir, star, model):
    simsize = 3
    #clear the window
    pl.clf()
    #set the x-axis
    pl.xlabel('EM Counts / 10^6')
    pl.ylabel('SNR')
    pl.plot(emcts/1e6, snr, 'bo', markersize=simsize)
    print "max snr: "+str(max(snr))
    pl.plot(emcts/1e6, model, 'ro', markersize=simsize)
    stddev = np.std(snr - model)
    print "Standard Deviation: "+str(stddev)
    pl.text(.55*max(emcts)/1e6, 0.375*max(snr), 'Polynomial Order: '+str(order))
    pl.text(.55*max(emcts)/1e6, 0.325*max(snr), 'Standard Deviation: {:.3}'.format(stddev))
    pl.text(.55*max(emcts)/1e6, 0.425*max(snr),'Star:'+str(order))
    #pl.show()
    pl.savefig(mdir+'projects/OTHER/NOTES/EXPM/em_cts_snr/'+star+'_emcts_snr_ord'+str(order)+'.eps')
```

```python
def model_em(emcts, snr, ord):
    z = np.polyfit(emcts, snr, ord)
    print z
    mod = np.zeros(len(emcts))
    for i in range(ord+1):
        mod = mod + z[ord - i] * emcts**i
    return mod
```

```python
def em_calib(sl):
    #to print the contents of the "OBJECT" tag:
    #sl.starlog.object
    snrall = sl.starlog.snrbp5500
    emavgstr = sl.starlog.emavg
    emnumsmpstr = sl.starlog.emnumsmp
    airmassall = sl.starlog.airmass
    #now use only the good ones, the values with velocities:
    gd = np.where(sl.starlog.cmnvel != 0)[0]
    #print gd
    snr = np.zeros(len(gd))
    emavg = np.zeros(len(gd))
    emnumsmp = np.zeros(len(gd))
    airmass = np.zeros(len(gd))
    #print len(gd)
    for i in range(len(gd)):
        #print i, airmassall[gd[i]]
        snr[i] = float(snrall[gd[i]])
        emavg[i] = float(emavgstr[gd[i]])
        emnumsmp[i] = float(emnumsmpstr[gd[i]])
        if str(airmassall[gd[i]]).strip() != 'airmass' and str(airmassall[gd[i]]).strip() != '':
            airmass[i] = float(airmassall[gd[i]])
    emcts = emavg * emnumsmp
    return emcts, snr, airmass
```

```python
def get_starlog(star, mdir):
    """PURPOSE: To return the starlog"""
    slogfn = mdir+'data/CHIRPS/starlogs/'+star+'log.dat'
    sl = readsav(slogfn)
    return sl
```

```python
def plotExpSnr(slg, savefig=False):
    mdnexptm = np.median(double(slg.exptime))
    mnexptm = np.mean(double(slg.exptime))

    f, ax = plt.subplots(2, sharex=True)
    ax[0].plot(slg.exptime, slg.snrbp5500, 'o', color='#6D92D2', markeredgecolor='#164BA7')
    #ax[0].set_xlabel('Exposure time [s]')
    ax[0].set_ylabel('SNR')
    ax[0].get_xaxis().tick_bottom()
    ax[0].yaxis.set_major_locator(MaxNLocator(prune='both'))
    ax[0].axvline(mdnexptm, color='#FC4F1A', alpha=0.75, lw=2.)
    ax[0].axvline(mnexptm, color='#FC4F1A', alpha=0.75, lw=2., ls=':')
    
    ax[1].hist(double(slg.exptime), bins=30, color='#6D92D2', edgecolor='#164BA7')
    ax[1].set_xlabel('Exposure Time [s]')
    ax[1].set_ylabel('# of Observations')
    ax[1].yaxis.set_major_locator(MaxNLocator(prune='both'))
    #ax[1].yaxis.set_major_locator(MaxNLocator(prune='lower'))
    print('The median exposure time is: {}'.format(mdnexptm))
    print('The mean exposure time is: {0}'.format(mnexptm))
    ax[1].axvline(mdnexptm, color='#FC4F1A', alpha=0.75, lw=2., label='Median')
    ax[1].axvline(mnexptm, color='#FC4F1A', alpha=0.75, lw=2., ls=':', label='Mean')
    ax[1].legend(loc='best')
    #ax[1].set_xlim([0, 1.1*double(np.max(slg.exptime))])
    
    # Fine-tune figure; make subplots close to each other 
    f.subplots_adjust(hspace=0)
    #x = plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)
    
    if (savefig is True):
        plt.savefig('ExposureSnr.eps')
    return mdnexptm
```

```python
def pltSnr(slg, mdnexptm, savefig=False):
    plt.hist(double(slg.snrbp5500), bins=30, color='#6D92D2', edgecolor='#164BA7')
    plt.xlabel('SNR')
    plt.ylabel('# of Observations')
    mdnsnr = np.median(double(slg.snrbp5500))
    mnsnr = np.mean(double(slg.snrbp5500))
    print('The median SNR: {}'.format(mdnsnr))
    print('The SNR standard deviation: {}'.format(np.std(double(slg.snrbp5500))))
    plt.axvline(mdnsnr, color='#FC4F1A', alpha=0.75, lw=2., label='Median')
    plt.axvline(mnsnr, color='#FC4F1A', alpha=0.75, lw=2., ls=':', label='Mean')
    plt.legend(loc='best')
    print('A is: {0}'.format(mdnexptm/(mdnsnr**2)))
    if (savefig is True):
        plt.savefig('histSNR.eps')

    return mdnsnr
```

```python
def plotExpSnrScatHist(slg, savefig=False):
    mdnexptm = np.median(double(slg.exptime))
    mnexptm = np.mean(double(slg.exptime))
    print('The median exposure time is: {}'.format(mdnexptm))
    print('The mean exposure time is: {0}'.format(mnexptm))

    mdnsnr = np.median(double(slg.snrbp5500))
    mnsnr = np.mean(double(slg.snrbp5500))
    print('The median SNR: {}'.format(mdnsnr))
    print('The SNR standard deviation: {}'.format(np.std(double(slg.snrbp5500))))

    f, ax = plt.subplots(2, 2)
    ax[1, 0].plot(slg.exptime, slg.snrbp5500, 'o', color='#6D92D2', markeredgecolor='#164BA7')
    ax[1, 0].set_xlabel('Exposure time [s]')
    ax[1, 0].set_xlim([0, 1.1*double(np.max(slg.exptime))])
    ax[1, 0].set_ylabel('SNR')
    ax[1, 0].get_xaxis().tick_bottom()
    ax[1, 0].yaxis.set_major_locator(MaxNLocator(prune='both'))
    ax[1, 0].axvline(mdnexptm, color='#FC4F1A', alpha=0.75, lw=2.)
    ax[1, 0].axvline(mnexptm, color='#FC4F1A', alpha=0.75, lw=2., ls=':')
    ax[1, 0].axhline(mdnsnr, color='#FC4F1A', alpha=0.75, lw=2.)
    ax[1, 0].axhline(mnsnr, color='#FC4F1A', alpha=0.75, lw=2., ls=':')
    
    ax[0, 0].hist(double(slg.exptime), bins=30, color='#6D92D2', edgecolor='#164BA7')
    ax[0, 0].yaxis.set_major_locator(MaxNLocator(prune='both'))
    ax[0, 0].set_xlim([0, 1.1*double(np.max(slg.exptime))])
    ax[0, 0].axvline(mdnexptm, color='#FC4F1A', alpha=0.75, lw=2., label='Median')
    ax[0, 0].axvline(mnexptm, color='#FC4F1A', alpha=0.75, lw=2., ls=':', label='Mean')
    ax[0, 0].set_ylabel('# of Observations')
    ax[0, 0].legend(loc='best')
    #x = plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)
    #y = plt.setp([a.get_yticklabels() for a in f.axes[:-1]], visible=False)
    
    #hide the x-ticks:
    ax[0, 0].get_xaxis().set_visible(False)
    
    #TOP-RIGHT QUADRANT:
    #hide the tick labels:
    ax[0, 1].get_xaxis().set_visible(False)
    ax[0, 1].get_yaxis().set_visible(False)
    ax[0, 1].axis('off')
    
    #BOTTOM-RIGHT QUADRANT:
    ax[1, 1].get_yaxis().set_visible(False)
    ax[1, 1].xaxis.set_major_locator(MaxNLocator(prune='both'))
    ax[1, 1].hist(double(slg.snrbp5500), bins=30, color='#6D92D2', 
                  orientation='horizontal', edgecolor='#164BA7')
    ax[1, 1].set_xlabel('# of Observations')
    ax[1, 1].axhline(mdnsnr, color='#FC4F1A', alpha=0.75, lw=2.)
    ax[1, 1].axhline(mnsnr, color='#FC4F1A', alpha=0.75, lw=2., ls=':')

    # Fine-tune figure; make subplots close to each other 
    f.subplots_adjust(hspace=0, wspace=0)
    
    if (savefig is True):
        plt.savefig('ExpSNRScatHist.eps')
    
    return mdnexptm, mdnsnr
```

There's a linear relation between exposure time and SNR$^{2}$:

$$t_{exp} = A*SNR^{2}$$

we can fill this in with the values for HD 20794 and solve for A:

$$A = \frac{t_{exp}}{SNR^{2}} = \frac{570.66}{237.8^{2}} = \frac{570.66}{56545.6} = 0.0101$$

```python
print(570.66/237.793134427**2)
print(mdnexptm/mdnsnr**2)
```

Now make arrays to store all this information. This will allow fitting for a B-V relation afterwards

```python
names = [20794]
spectypes = ['G8V']
vmags = np.array(4.26)
bvarr = np.array(0.71)
coefs = np.array(mdnexptm/mdnsnr**2)
exptimes = np.array(mdnexptm)
snrs = np.array(mdnsnr)
```

This makes the full exposure time equation:

$$t_{exp} = 1.9 \cdot 10^{-4} (SNR)^{2} 10^{0.4V}$$

where V is the apparent magnitude and SNR is the signal-to-noise ratio.

```python
def expcalc(snr, vmag):
    return 0.000189646878259 * snr**2 * 10**(0.4*vmag)
```

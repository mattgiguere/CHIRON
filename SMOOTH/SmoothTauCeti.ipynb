### 3-day Average Filter

What I really want is a filter that averages over a specified time period, not the number of points. I'm going to create that now.

```python
from numba import jit

@jit
def runningDayAverage(time, value, window):
    outpt = np.zeros(len(time))
    for i in range(len(time)):
        x = np.where((time > (time[i] - window/2.)) & (time < (time[i] + window/2.) ))[0]
        outpt[i] = np.sum(value[x])/len(x)
    return outpt
```

```python
daysmooth2 = runningDayAverage(jds, mnvels, 2.)
daysmooth3 = runningDayAverage(jds, mnvels, 3.)
daysmooth4 = runningDayAverage(jds, mnvels, 4.)
```
